```python
from Util.Problems import Problem, solution


class P001(Problem):
    number = 1
    title = "Multiples of 3 or 5"
    description ="""<p>If we list all the natural numbers below $10$ that are multiples of $3$ or $5$, we get $3, 5, 6$ and $9$. The sum of these multiples is $23$.</p>
<p>Find the sum of all the multiples of $3$ or $5$ below $1000$.</p>"""
```

```python
p = P001()
p.describe()
```

### Solution notes
First we loop through all numbers from $1-1000$, if they are multiples of 3 or 5, add them to the total.

```python
@solution(P001, first=True, make_fast=True)
def simple():
    sum_of_multiples = 0
    for i in range(1, 1000):
        if i % 3 == 0 or i % 5 == 0:
            sum_of_multiples += i
    return sum_of_multiples
```

```python
p.test_all()
```

We now approach it from the opposite direction, we loop through all multiples of $3$, add them to the sum, then loop through the multiples of $5$, check if they are not a multiple of $3$ to prevent duplicates and add those. This prevents the expensive loop over all numbers.

```python
@solution(P001, best=True, make_fast=True)
def multiples():
    sum_of_multiples = 0
    i = 3
    while i < 1000:
        sum_of_multiples += i
        i += 3
    j = 5
    while j < 1000:
        if j % 3 != 0:
            sum_of_multiples += j
        j += 5
    return sum_of_multiples
```

```python
p.test_all()
```
