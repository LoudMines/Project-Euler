```python
from Util import Problem, solution

class P000(Problem):
    number = 0
    title = "Problem Zero"
    description = (
        "A number is a perfect square, or a square number, if it is the square of a positive integer.\n" +
        "For example, $25$ is a square number because $5^2 = 5 \\times 5 = 25$; it is also an odd square.\n\n" +
        "The first $5$ square numbers are: $1, 4, 9, 16, 25$, and the sum of the odd squares is $1 + 9 + 25 = 35$.\n\n" +
        "Among the first n (747 thousand) square numbers, what is the sum of all the odd squares?"
    )
```

```python
p = P000()
p.describe()
```

### Solution notes
Loop through all numbers $1-n$, square them and if the square is odd, add it to the total.

```python
@solution(P000, first=True, make_fast=True)
def brute_force():
    n = 747000
    squares_sum = 0
    for i in range(1, n + 1):
        square = i ** 2
        if square % 2 != 0:
            squares_sum += square
    return squares_sum
```

```python
p.test_all()
```

For $i^2$ to be odd, $i$ itself must be odd, so we just loop through the odd numbers, square and add those instead.

```python
@solution(P000, best=True, make_fast=True)
def odd_n_squares():
    n = 747000
    squares_sum = 0
    for i in range(1, (n + 2) // 2):
        odd_n = (2 * i) - 1
        squares_sum += odd_n ** 2
    return squares_sum
```

```python
p.test_all()
```
