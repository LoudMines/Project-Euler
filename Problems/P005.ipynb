```python
from Util.Problems import Problem, solution
from Util.math_functions import is_prime

class P005(Problem):
    number = 5
    title = "Smallest Multiple"
    description = """<p>$2520$ is the smallest number that can be divided by each of the numbers from $1$ to $10$ without any remainder.</p>
<p>What is the smallest positive number that is <strong class="tooltip">evenly divisible<span class="tooltiptext">divisible with no remainder</span></strong> by all of the numbers from $1$ to $20$?</p>"""
```

```python
p = P005()
p.describe()
```

### Solution notes
We start by the very simplistic method of increasing a number n until we find an n which has the numbers $1$ to $20$ as factors.

```python
@solution(P005, first=True, make_fast=True)
def brute_force():
    # The number is bigger than the smallest number which can be divided by 1-10, so we use that as
    # an initial n
    n = 2520
    factors = [f for f in range(2, 21)]
    while True:
        divider = True
        for f in factors:
            if n % f != 0:
                divider = False
                break
        if divider:
            return n
        n += 1
```

```python
p.test_all(repeats=1)
```

The number we are looking for can be found by multiplying all prime factors between 1-20. Some of these prime factors need to be multiplied multiple times however, E.G: $8 = 2^3$, so we create a dict of the prime factors, then for all non-primes we check their prime factorisation and if they require a prime more than once, we update the count of that prime in the dict.

```python
@solution(P005, best=True, make_fast=False)
def prime_factorisation():
    factors = [f for f in range(2, 21)]
    prime_factors = {f: 1 for f in factors if is_prime(f)}
    # From simple tests, this seems to be faster than checking if f is in prime_factors
    non_prime_factors = [f for f in factors if not is_prime(f)]
    for n in non_prime_factors:
        prime_iterator = iter(prime_factors)
        current_prime_factor = next(prime_iterator)
        current_prime_count = 0
        while n != 1:
            if n % current_prime_factor == 0:
                current_prime_count += 1
                n //= current_prime_factor
                if current_prime_count > prime_factors[current_prime_factor]:
                    prime_factors[current_prime_factor] = current_prime_count
            else:
                current_prime_count = 0
                current_prime_factor = next(prime_iterator)

    result = 1
    for prime_factor, count in prime_factors.items():
        result *= prime_factor ** count

    return result
```

```python
p.test_all(repeats=10)
```
