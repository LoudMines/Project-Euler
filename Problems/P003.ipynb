```python
import math

from Util import Problem, solution
from math_functions import is_prime, sieve_eratosthenes

class P003(Problem):
    number = 3
    title = "Largest Prime Factor"
    description = (
        "<p>The prime factors of $13195$ are $5, 7, 13$ and $29$.</p>" +
        "<p>What is the largest prime factor of the number $600851475143$?</p>"
    )
```

```python
p = P003()
p.describe()
```

### Solution notes
First we divide the number by all numbers and save the biggest one we find. We are checking non-primes as well, but this is not an issue, as all smaller primes have already been checked, so these do not divide into the remainder and thus will not affect the output.

```python
@solution(P003, first=True, make_fast=True)
def brute_force():
    i = 2
    remainder = 600851475143
    largest_prime_factor = 0
    while True:
        if remainder % i == 0:
            remainder /= i
            largest_prime_factor = i
        else:
            i += 1
        if remainder == 1:
            break
    return largest_prime_factor
```

```python
p.test_all()
```

The number $600851475143$ is odd, therefore no even numbers need to be tested

```python
@solution(P003, best=True, make_fast=True)
def brute_force_no_even():
    i = 3
    remainder = 600851475143
    largest_prime_factor = 0
    while True:
        if remainder % i == 0:
            remainder /= i
            largest_prime_factor = i
        else:
            i += 2
        if remainder == 1:
            break
    return largest_prime_factor
```

```python
p.test_all()
```

We now test if numbers are prime, if they are not, we can skip them and continue. This turns out to be more computationally expensive though.

```python
@solution(P003, make_fast=True)
def primes_only():
    i = 3
    remainder = 600851475143
    largest_prime_factor = 0
    while True:
        if not is_prime(i):
            i += 2
            continue
        if remainder % i == 0:
            remainder /= i
            largest_prime_factor = i
        else:
            i += 2
        if remainder == 1:
            break
    return largest_prime_factor
```

```python
p.test_all()
```

We now use a prime sieve to only consider primes. This turns out to be much more expensive. Might be better with a faster sieve. Worth testing in the future once I implement a better sieve.

```python
@solution(P003, make_fast=True)
def prime_sieve():
    n = 600851475143
    # retrieves a list of all prime numbers from 2 to the square root of n
    prime_options = sieve_eratosthenes(math.ceil(math.sqrt(n)))
    prime_options.reverse()
    for prime in prime_options:
        if n % prime == 0:
            return prime
    return None
```
