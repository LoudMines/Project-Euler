```python
from Util.Problems import Problem, solution
from math_functions import sieve_eratosthenes

class P007(Problem):
    number = 7
    title = "10 001st Prime"
    description = """<p>By listing the first six prime numbers: $2, 3, 5, 7, 11$, and $13$, we can see that the $6$th prime is $13$.</p>
<p>What is the $10\\,001$st prime number?</p>"""
```

```python
p = P007()
p.describe()
```

### Solution notes
Prime sieves are algorithms which generate all primes up to a chosen limit. In this implementation, a basic sieve is chosen with an arbitrary limit, which gets increased if the answer is not found.

```python
@solution(P007, first=True, make_fast=True)
def arbitrary_limit():
    # 2^16 is chosen as an arbitrary start value. When the 10.001st prime is not found, the power of 2 is increased by 1
    sieve_limit = 16
    while True:
        sieve_result = sieve_eratosthenes(2 ** sieve_limit)
        if len(sieve_result) >= 10001:
            return sieve_result[10000]
        sieve_limit += 1
```

```python
p.test_all()
```

Prime-counting functions are functions which count the number of primes less than or equal to a real number $x$. These functions are denoted by $\pi(x)$ (unrelated to the number $\pi$).

These functions are quite complex, so I will not implement one here, however it is known that

$\frac{x}{\log(x-1)} < \pi(x)$

Therefore, if $\frac{x}{\log(x-1)} > 10,001$, we know the prime we are looking for is smaller than $x$, and we can use it as a non-arbitrary limit for the sieve. Using a prime-counting function could further optimise this.

The result of the equality $\frac{x}{\log(x-1)} = 10,001$ is $x ≈ 116683.812494882...$, so as an upper limit for the sieve, we will use $lim = 116684$.

```python
@solution(P007, best=True, make_fast=True)
def calculated_limit():
    # 2^16 is chosen as an arbitrary start value. When the 10.001st prime is not found, the power of 2 is increased by 1
    sieve_limit = 116684
    sieve_result = sieve_eratosthenes(sieve_limit)
    return sieve_result[10000]
```
